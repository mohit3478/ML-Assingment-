##***A python program for import random numbers with an range and also get the base_frequency, highest_base, realtive_frequency and plot a graph between relative_frequency and highest_base.***

```python
import random # We import random for getting diffrent random numbers.
import matplotlib.pyplot as plt # Impoer this library for plotting an mathematical graph between two data.
```

##At first create an dictionary name mapping_dict and take an range for dictionary keys and values.

```python
mapping_dict = dict(zip(range(0,10),range(0,10))) # This mapping_dict is created for containing all the numbers(0-9) and alphabate(A-Z).
```

```python
mapping_dict # The created dictionary is in output with no character value.
```

```python
mapping_dict.keys()
#here is a seperate keys dictionary
```

```python
mapping_dict.values()
# here is a seperate values dictionary
```

##For getting random characters  values, we have make each values ponit to string,

```python
for k in mapping_dict.keys():

  mapping_dict[k] = str(mapping_dict[k])
```

```python
mapping_dict
```

```python
mapping_dict[10] = chr(65) #
```

```python
mapping_dict
```

```python
for i in range(0,26):
  mapping_dict[10+i] = chr(65+i)
```

```python
 mapping_dict # Here dictionary mapping_dict contain all the numbers(0-9) and alphabates(A-Z).
```

```python
# for making mapping_dict reverse.
reverse_mapping_dict = dict()
for k in mapping_dict.keys():
  reverse_mapping_dict[mapping_dict[k]] = k
```

```python
reverse_mapping_dict
```

##Now for import random numbers.

```python
sample_size = 10000 # for give an range of importing numbers.

random_numbers = list() # making an empty list

for i in range(0,sample_size): # for loop for given range
  blank_str = str() #making an empty string for all string values.

  random_digit = random.randint(0,35) # import an random number.#
  blank_str = blank_str + mapping_dict[random_digit]

  random_digit = random.randint(0,35) # import another random number.
  blank_str = blank_str + mapping_dict[random_digit]


  for i in range(0,sample_size):
    coin_toss = random.randint(0,1)
    if coin_toss == 1:
      random_digit = random.randint(0,35)
      blank_str = blank_str + mapping_dict[random_digit]

    else:
      break
    random_numbers.append(blank_str)
```

```python
random_numbers
```

##After importing random numbers we have to the highest base of that number.
##So,

```python
for balnk_str in random_numbers:
  highest_digit = max(blank_str)
  highest_base = reverse_mapping_dict[highest_digit] + 1
  print("Highest Base of {} is {}".format(blank_str,highest_base))
```

```python
# for getting relative frequency of random numbers by there highest base.
base_frequency = dict()

for blank_str in random_numbers:
  highest_digit = max(blank_str)
  highest_base = reverse_mapping_dict[highest_digit] + 1

  if highest_base in base_frequency.keys():
    base_frequency[highest_base] += (1/sample_size)
  else:
    base_frequency[highest_base] = (1/sample_size)
```

```python
base_frequency
```
